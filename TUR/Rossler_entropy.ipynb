# 決定論的力学系における熱力学的不確定性

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def dz_dt(v, t, a,b,c):
    x,y,z= v
    
    dx = -y-z
    dy = x + a*y
    dz = b+x*z-c*z 
    
    return np.stack([dx,dy,dz])
```

```python
a=0.2
b=0.2
c=5.7

v_init = np.array([1.0, 0.0, 0.0])

time_true = np.arange(700).astype(float)

#v = odeint(dz_dt, v_init, time_true, a,b,c, rtol=1e-7, atol=1e-6, mxstep=10000)
v = odeint(dz_dt, v_init, time_true, (a,b,c,))
```

```python
plt.figure(figsize=(15,10))
plt.plot(v, label=['x', 'y', 'z'])
plt.xlabel('time[s]')
plt.ylabel('value')
plt.legend()
plt.show()
```

## 化学反応系

```python
#Minimal Willamowski–Rössler model (MWR model)

k1=30
k2=1
k3=10
k4=1
k5=16.5
km5=0.5
km1=0.25

def dz_dt(v, t, k1,k2,k3,k4,k5,km5,km1):
    x1,x2,x3= v

    dx1 = k1*x1-km1*x1*x1-k2*x1*x2-k4*x1*x3
    dx2 = k2*x1*x2-k3*x2
    dx3 = -k4*x1*x3+k5*x3-km5*x2
    
    return np.stack([dx1,dx2,dx3])
```

```python
v_init = np.array([1.0, 0.0, 0.0])

time_true = np.arange(700).astype(float)

v = odeint(dz_dt, v_init, time_true, (k1,k2,k3,k4,k5,km5,km1))

plt.figure(figsize=(15,10))
plt.plot(v, label=['x', 'y', 'z'])
plt.xlabel('time[s]')
plt.ylabel('value')
plt.legend()
plt.show()
```
