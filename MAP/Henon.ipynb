# Henon map

$x=1+y-ax^2$

$y=bx$

```python
def Henonmap(a,b,T,x=0.1,y=0.1,M=1e6):
    r=lambda x:max(min(x,M),-M)
    xs=[]
    ys=[]
    for t in range(T):
        xn=1+y - a*x*x
        yn= b*x
        x,y=xn,yn
        xs.append(r(x))
        ys.append(r(y))
        if(abs(x)>M or abs(y)>M):
            break
    return xs,ys

def full_plot(T):
    for a in range(1,10):
        for b in range(10):
            xs,ys=Henonmap(a/10,b/10,T)
            plt.scatter(xs,ys,marker=".")    
```

```python
a=1.4
b=0.3
T=400
xs,ys=Henonmap(a,b,T,0.,0)

plt.scatter(xs,ys,marker=".") 
plt.show()
plt.plot(xs)
plt.plot(ys)
```

```python
a=1.4
b=0.3
T=400
xs,ys=Henonmap(a,b,T,0.6,0.5)

plt.scatter(xs,ys,marker=".") 
plt.show()
plt.plot(xs)
plt.plot(ys)
```

# crisis

```python
T=100
b=0.3
M=1e6
for a in np.arange(0.,1.5, 0.1):
    xs,ys=Henonmap(a,b,T,0.6,0.5,M=M)
    
    fig = plt.figure(figsize=(16,8))
    fig.suptitle("a={},b={}".format(a,b))
    ax = fig.add_subplot(1, 2, 1)
    ax.scatter(xs, ys,marker=".") 
    bx = fig.add_subplot(1, 2, 2)
    bx.plot(xs)
    
    fig.tight_layout()
    fig.show()
    if(abs(xs[-1])>M or abs(ys[-1])>M):
        break
```
