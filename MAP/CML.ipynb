```python
import numpy as np
import matplotlib.pyplot as plt
import math
from copy import deepcopy
```

```python
def CML1D(N,T,a,eps,seed=1):
    rng = np.random.default_rng(seed)
    x=np.random.rand(N)
    xn=deepcopy(x)
    for t in range(T):
        x=1-a*x*x
        xl=np.concatenate([[x[-1]],x[:-1]])
        xr=np.concatenate([x[1:],[x[0]]])
        x=(1-eps)*x+eps/2*(xl+xr)
        xn=np.concatenate([xn,x])
    xn=xn.reshape(N,T+1)
    return xn
```
