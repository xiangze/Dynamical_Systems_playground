# Heteroclinic cycle

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from copy import deepcopy
```

# Chawanya

```python
def gamedynamics(T,w,N=5,seed=1):
    rng = np.random.default_rng(seed)
    #x=np.random.rand(N)
    x=np.array([0.2]*N)
    xs=deepcopy(x)
    
    for t in range(T):
        s=np.dot(w,x)
        sb=np.dot(s,x)
        xn= x*(s-sb)
        x=xn
        xs=np.concatenate([xs,x])
    xs=xs.reshape(N,T+1)
    return xs
   
X=0.32                
w=[[ 1.0 ,-20.0, - 0.4, - 1.0,  1.0 ],
    [1.5, 0.0, -0.7,- 7.3,  0.5  ],
    [X, 1.0,   0.0,   0.0,- 0.1  ],
    [0.9,  0.8,   1.0,- 1.0, - 0.1 ],
    [0.0, - 8.0,  0.7 ,  1.3,   0.0 ]]
w=np.array(w)
```

```python
xs=gamedynamics(100,w)
```
